Source: https://data.wprdc.org/dataset/2020-census-redistricting-data-extracts/resource/a8414ed5-c50f-417e-bb67-82b734660da6. My partner and I are both Asian, and we want to live in a neighborhood with more Asian people. So I will use the data to find the Asian ratio in different neighborhoods.

```python
#import
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
population = pd.read_csv("population.csv")
population
```

```python
myDict = dict()
for x in range(len(population)):
    key = population.loc[x]["Neighborhood"]
    myDict[key] = (population.loc[x]["2020_Asian_Alone_Population"] / population.loc[x]["2020_Total_Population"])
myDict
```

```python
RatioSeries = pd.Series(myDict, name = "Fields")
RatioSeries
```

```python
RatioSeries = RatioSeries.sort_values(ascending=False).head(30)
RatioSeries.plot.bar()
```
