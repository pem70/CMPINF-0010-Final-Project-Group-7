Menghao's Jupyter NB: park meter in Neighborhoods

Source: https://data.wprdc.org/dataset/pittsburgh-parking-meters-and-payment-points. My partner and I wanted to travel by car, so we wanted to stay in an area with easy parking. I ended up picking the ten places with the most parking meters.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
permitparking = pd.read_csv('permitparking.csv')
permitparking
```

```python
district_counts = permitparking.groupby('neighborhood')['district'].count().reset_index(name='count')
district_counts.sort_values(by='count', ascending=False, inplace=True)
most_permitparking_zone = district_counts.iloc[0]['neighborhood']
#get top 10 zone with most park point
top30 = district_counts.head(30)
print(top30)
```

```python

# create the plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top30['neighborhood'], top30['count'], align='center')

# set labels and title
ax.set_xlabel('Number of Park Points')
ax.set_ylabel('district')
ax.set_title('Top 30 Zones with Most Park Points')

# show the plot
plt.show()
```

Peizhao's Jupyter NB: Asian Ratio

Source: https://data.wprdc.org/dataset/2020-census-redistricting-data-extracts/resource/a8414ed5-c50f-417e-bb67-82b734660da6. My partner and I are both Asian, and we want to live in a neighborhood with more Asian people. So I will use the data to find the Asian ratio in different neighborhoods.



```python
population = pd.read_csv("population.csv")
population
```

```python
ADict = dict()
for x in range(len(population)):
    key = population.loc[x]["Neighborhood"]
    ADict[key] = (population.loc[x]["2020_Asian_Alone_Population"] / population.loc[x]["2020_Total_Population"])
ADict
```

```python
RatioSeries = pd.Series(ADict, name = "Fields")
RatioSeries
```

```python
RatioSeries = RatioSeries.sort_values(ascending=False).head(30)
RatioSeries.plot.bar()
```

### Peizhao's Jupyter NB: Arrest in Neighborhoods

Source: https://data.wprdc.org/dataset/arrest-data. My partner and I want to live in a safe place. I use the arrest data to analyze which neighborhood is relatively safe. However, small neighborhoods with few population have low arrest incidents that affect the analysis. Thus, the final result only analyze the major neighborhood in Pittsburgh city. 

```python
#read file
arrest = pd.read_csv("Arrest report.csv")
arrest["INCIDENTNEIGHBORHOOD"]
```

```python
#create dictionary
SDict = dict()
for i in range(len(arrest)):
    temp = arrest.loc[i, "INCIDENTNEIGHBORHOOD"]
    if(temp not in SDict):
        SDict[temp] = 1
    else:
        SDict[temp] +=1
# Divided by population
for keys in SDict:
    for x in range(len(population)):
        if keys == population.loc[x]["Neighborhood"]:
            SDict[keys] = ( SDict[keys] / population.loc[x]["2020_Total_Population"])
    # Deal with error data when arrest neighbourhood name does not match related population neighbourhood name
    if SDict[keys] >= 10:
        SDict[keys] = 10

SDict["Central Business District"]
```

```python
#convert to series
arrestSeries = pd.Series(SDict, name = "Fields")
arrestSeries
```

```python
#Get the first 30 
ArrestSeriesInOrder = arrestSeries.sort_values(ascending=True).head(30)
ArrestSeriesInOrder
```

```python
#Plot
ArrestSeriesInOrder.plot.bar()
```
