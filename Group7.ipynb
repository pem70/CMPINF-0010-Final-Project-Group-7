## Park meter in Neighborhoods

By Menghao<br>

Source: https://data.wprdc.org/dataset/residential-permit-parking-program-areas. My partner and I wanted to travel by car, so we wanted to stay in an area with easy parking. I ended up picking the ten places with the most parking meters.

```python
#import pandas and plot

import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# read dataset
permitparking = pd.read_csv('permitparking.csv')
permitparking
```

```python
district_counts = permitparking.groupby('neighborhood')['district'].count().reset_index(name='count')
district_counts.sort_values(by='count', ascending=False, inplace=True)
most_permitparking_zone = district_counts.iloc[0]['neighborhood']
#get top 10 zone with most park point
top30 = district_counts.head(30)
print(top30)
```

```python

# create the plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top30['neighborhood'], top30['count'], align='center')

# set labels and title
ax.set_xlabel('Number of Park Points')
ax.set_ylabel('neighborhood')
ax.set_title('Top 30 Zones with Most Park Points')

# show the plot
plt.show()
```

## Asian Ratio

By Peizhao<br>

Source: https://data.wprdc.org/dataset/2020-census-redistricting-data-extracts/resource/a8414ed5-c50f-417e-bb67-82b734660da6. This dataset records population of different races in each neighborhood. My partner and I are both Asian, and we want to live in a neighborhood with more Asian people. So I will use the dataset to find the Asian ratio in different neighborhoods.



```python
# read dataset
population = pd.read_csv("population.csv")
population
```

```python
#create dictionary
ADict = dict()
for x in range(len(population)):
    key = population.loc[x]["Neighborhood"]
    # get Asian Ratio by (Asian/Total)
    ADict[key] = (population.loc[x]["2020_Asian_Alone_Population"] / population.loc[x]["2020_Total_Population"])
ADict
```

```python
#Convert to Series
RatioSeries = pd.Series(ADict, name = "Fields")
RatioSeries
```

```python
#Sort and plot
RatioSeries = RatioSeries.sort_values(ascending=False).head(30)
RatioSeries.plot.barh()
```

## Arrest in Neighborhoods
By Peizhao<br>

Source: https://data.wprdc.org/dataset/arrest-data. My partner and I want to live in a safe place. This dataset records each arrest report and its lacation. I will collect the data to find out which neighborhood has the lowest arrest rate (arrest report per person). Low arrest rate shows that the neighborhood is relatively safe in most cases.

```python
#read file
arrest = pd.read_csv("Arrest report.csv")
arrest["INCIDENTNEIGHBORHOOD"]
```

```python
#create dictionary
SDict = dict()
for i in range(len(arrest)):
    #count total arrest reports for each neighborhood
    temp = arrest.loc[i, "INCIDENTNEIGHBORHOOD"]
    if(temp not in SDict):
        SDict[temp] = 1
    else:
        SDict[temp] +=1
# Divided by population
for keys in SDict:
    for x in range(len(population)):
        if keys == population.loc[x]["Neighborhood"]:
            SDict[keys] = ( SDict[keys] / population.loc[x]["2020_Total_Population"])
    # Deal with error data when arrest neighbourhood name does not match related population neighbourhood name
    if SDict[keys] >= 10:
        SDict[keys] = 10
```

```python
#convert to series
arrestSeries = pd.Series(SDict, name = "Fields")
arrestSeries
```

```python
#Get the first 30 
ArrestSeriesInOrder = arrestSeries.sort_values(ascending=True).head(30)
ArrestSeriesInOrder
```

```python
#Plot
ArrestSeriesInOrder.plot.barh()
```

## Combining submetrics
Description: We combine sub-metrics using a point-based system.

### Traffic (Parking)
Top 5: 5 points <br>
6 - 10: 4 points <br>
11 - 15: 3 points <br>
16 - 20: 2 points <br>
21 - 30: 1 point <br>

### Culture (Asian Ratio)

### Safety (Arrest rate)

21 - 30: 1 point <br>

```python
myDict = dict()
#Get rank
count = 0
for indx,value in RatioSeries.items():
    #Distribute points
    if count < 5:
        points = 5
    elif count < 10:
        points = 4
    elif count < 15:
        points = 3
    elif count < 20:
        points = 2
    else:
        points = 1
    
    temp = indx
    #Calculate points awarded
    if(temp not in myDict):
        myDict[temp] = points
    else:
        myDict[temp] += points
    
    count += 1
#Reset
count = 0

for indx,value in ArrestSeriesInOrder.items():
    #Distribute points
    if count < 5:
        points = 5
    elif count < 10:
        points = 4
    elif count < 15:
        points = 3
    elif count < 20:
        points = 2
    else:
        points = 1
    
    temp = indx
    #Calculate points awarded
    if(temp not in myDict):
        myDict[temp] = points
    else:
        myDict[temp] += points
    
    count += 1
#Reset    
count = 0

for x in top30["neighborhood"]:
    #Distribute points
    if count < 5:
        points = 5
    elif count < 10:
        points = 4
    elif count < 15:
        points = 3
    elif count < 20:
        points = 2
    else:
        points = 1
    #Convert neighborhood format
    temp = x.title()
    #Calculate points awarded
    if(temp not in myDict):
        myDict[temp] = points
    else:
        myDict[temp] += points
    
    count += 1
#Convert to Series and visualize
PointSeries = pd.Series(myDict, name = "Fields").sort_values(ascending=False).head(30)
PointSeries.plot.barh()
```
