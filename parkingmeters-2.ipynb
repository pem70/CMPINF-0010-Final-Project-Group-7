Source: https://data.wprdc.org/dataset/pittsburgh-parking-meters-and-payment-points.My partner and I wanted to travel by car, so we wanted to stay in an area with easy parking. I ended up picking the ten places with the most parking meters.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

parkingmeters = pd.read_csv('parkingmeters.csv')
parkingmeters
```

```python
zone_counts = parkingmeters.groupby('zone')['id'].count().reset_index(name='count')
zone_counts.sort_values(by='count', ascending=False, inplace=True)
most_parkingmeters_zone = zone_counts.iloc[0]['zone']
#get top 10 zone with most park point
top10 = zone_counts.head(10)
print(top10)
```

```python
zone = parkingmeters.groupby('zone')
print(zone)
```

```python
import matplotlib.pyplot as plt

# create the plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top10['zone'], top10['count'], align='center')

# set labels and title
ax.set_xlabel('Number of Park Points')
ax.set_ylabel('Zone')
ax.set_title('Top 10 Zones with Most Park Points')

# show the plot
plt.show()
```

```python
#import
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#read file
arrest = pd.read_csv("Arrest report.csv")

arrest["INCIDENTNEIGHBORHOOD"]
```

```python
#create dictionary
myDict = dict()
for i in range(len(arrest)):
    temp = arrest.loc[i, "INCIDENTNEIGHBORHOOD"]
    if(temp not in myDict):
        myDict[temp] = 1
    else:
        myDict[temp] +=1

myDict["Squirrel Hill South"]
```

```python
#convert to series
arrestSeries = pd.Series(myDict, name = "Fields")
arrestSeries
```

```python
#Get the first 30 to exclude small neighborhoods.
SeriesInOrder = arrestSeries.sort_values(ascending=False).head(30)
#Arrange in ascending order.
SeriesInOrder = SeriesInOrder.sort_values(ascending=True)
SeriesInOrder
```

```python
#Plot
SeriesInOrder.plot.bar()
```

Peizhao's Jupyter NB: Asian Ratio

Source: https://data.wprdc.org/dataset/2020-census-redistricting-data-extracts/resource/a8414ed5-c50f-417e-bb67-82b734660da6. My partner and I are both Asian, and we want to live in a neighborhood with more Asian people. So I will use the data to find the Asian ratio in different neighborhoods.



```python
#import
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
population = pd.read_csv("population.csv")
population
```

```python
myDict = dict()
for x in range(len(population)):
    key = population.loc[x]["Neighborhood"]
    myDict[key] = (population.loc[x]["2020_Asian_Alone_Population"] / population.loc[x]["2020_Total_Population"])
myDict
```

```python
RatioSeries = pd.Series(myDict, name = "Fields")
RatioSeries
```

```python
RatioSeries = RatioSeries.sort_values(ascending=False).head(30)
RatioSeries.plot.bar()
```
