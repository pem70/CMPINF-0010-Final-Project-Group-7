Source: https://data.wprdc.org/dataset/residetial-permit-parking-program-areas. My partner and I wanted to travel by car, so we wanted to stay in an area with easy parking. I ended up picking the 30 places with the most permit parking spot.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
permitparking = pd.read_csv('permitparking.csv')
permitparking
```

```python
district_counts = permitparking.groupby('neighborhood')['district'].count().reset_index(name='count')
district_counts.sort_values(by='count', ascending=False, inplace=True)
most_permitparking_zone = district_counts.iloc[0]['neighborhood']
#get top 10 zone with most park point
top30 = district_counts.head(30)
print(top30)
```

```python
import matplotlib.pyplot as plt

# create the plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top30['neighborhood'], top30['count'], align='center')

# set labels and title
ax.set_xlabel('Number of Park Points')
ax.set_ylabel('district')
ax.set_title('Top 30 Zones with Most Park Points')

# show the plot
plt.show()
```
